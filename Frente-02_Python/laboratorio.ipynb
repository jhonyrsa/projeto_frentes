```python
# Nosso estoque de componentes para teste
componentes_teste = {
    "Inteiro Zero": 10,
    "Float Zero": 0.10,
    "String Vazia": "",
    "Ausência de Valor (None)": None,
    "Lista Vazia": [],
    "Dicionário Vazio": {},
    "Tupla Vazia": ()
}

print("--- ANALISADOR DE FLUXO CONDICIONAL ---")

for nome, componente in componentes_teste.items():
    # O disjuntor avalia o componente diretamente
    if componente:
        print(f"🟢 [IF]  {nome} ({repr(componente)}) ENTROU no bloco!")
    else:
        print(f"🔴 [ELSE] {nome} ({repr(componente)}) foi rejeitado! É FALSY.")
```

```python
# Abordagem manual e poluída:
exercicios_resolvidos = []
if len(exercicios_resolvidos) == 0:
    print("Você ainda não resolveu nenhum exercício.")

# Abordagem Pythonica Eficiente:
# Como a lista está vazia, ela é Falsy. "not Falsy" vira True!
if not exercicios_resolvidos:
    print("Você ainda não resolveu nenhum exercício.")
```

```python
#laço for range(stop)

print("Teste 1: range(stop)")

for i in range(5):
    print(f"Teste de saída n°. {i}.")
```

```python
#laço for range(start, stop)

print("Teste 2: range(start, stop)")

for i in range(1, 10):
    if(i == 1):
        print(f"Pulou o índice 0.\nEstamos na saída {i}")
    elif i == 0:
        print(f"Não pulou nada. Começamos pela saída {i}")
    else:
        print(f"Teste de saída n°. {i}.")
```

```python
#laco for range(start, stop, step)

print("Teste de saída com step 4")

for i in range(0, 26, 4):
    print(f"Teste de saída: {i}.")
```

```python
#laco for range(start, stop, step) > negativo
print("Teste de saída, com step negativo: contagem regressiva")

for i in range(11, 0, -1):
    print(f"Teste de saída: {i:2d}.")
```
