```python
#valor se verdadeiro if condicao else valor se faso

usuario_conectado = True

print("Acesso liberado ao Servidor." if usuario_conectado else "Acesso negado: Chave SSH inválida")
```

```python
consumo_kwh = 450.0

print(f"Consumo atual: {consumo_kwh} | Classificação: {'Alto' if consumo_kwh > 400 else 'Normal'}")
```

```python
codigo_problema = 1005

print(f'O problema {codigo_problema} está na esteira de códigos {"PARES" if codigo_problema % 2 == 0 else "IMPARES"}')
```
