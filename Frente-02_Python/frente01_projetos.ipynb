##### Calculadora de Dados Pessoais Interativa
Crie um programa Python que interaja com o usuário via terminal usando apenas print() e input().
O programa deve:

(1) solicitar nome, idade e altura (em metros) do usuário;

(2) calcular e exibir: ano de nascimento aproximado, altura em centímetros, quantos anos faltam para os 100 anos, iniciais do nome;

(3) exibir um "cartão de apresentação" formatado com todas as informações coletadas e calculadas;

(4) tratar erros básicos de conversão com mensagens claras ao usuário (ex: digitar letras onde se espera número). Usar apenas print(), input() e operações aritméticas básicas — sem if/else ainda.

```python
# (1) solicitar nome, idade e altura (em metros) do usuário;
nome = input("Informe seu primeiro nome: ")
idade = int(input("Informe sua idade: "))
altura_m = float(input("Informe sua altura em metros: "))

# (2) calcular e exibir: ano de nascimento aproximado, altura em centímetros,
# quantos anos faltam para os 100 anos, iniciais do nome;
ANO_ATUAL = 2026
ano_nasc = ANO_ATUAL - idade
altura_cm = int(altura_m * 100)
faltam_cem = 100 - idade
inicial_nome = nome[0]

print(f"Ano de nascimento: {ano_nasc}")
print(f"Altura (em centímetros): {altura_cm}")
print(f"Faltam para 100 anos: {faltam_cem}")
print(f"Iniciais do nome: {inicial_nome}")

# (3) exibir um "cartão de apresentação" formatado com todas as informações coletadas e calculadas;
print("\n\nCartão de apresentação")
print(f"Nome: {nome}")
print(f"Inicial do nome {inicial_nome}")
print(f"Idade: {idade}")
print(f"Ano de nascimento: {ano_nasc}")
print(f"Altura: {altura_m}")
print(f"Altura: {altura_cm}")
print(f"Faltam para 100 anos: {faltam_cem}")

```

##### Calculadora Científica com Tipos e Operadores

(1) solicitar dois números ao usuário com input() e converter para float;

(2) calcular e exibir com print() todos os resultados: soma, subtração, multiplicação, divisão, divisão inteira, módulo, potenciação;

(3) exibir o tipo de cada resultado com type();

(4) calcular e exibir: se o primeiro número é maior, menor ou igual ao segundo (usando comparadores e exibindo o resultado bool);
(5) calcular manipulações de bits (AND, OR, XOR) convertendo para int;

(6) demonstrar conversões: mostrar o primeiro número como int, float, str e bool;

(7) usar type hints em todas as variáveis. Sem if/else ainda — usar apenas operadores e print() para mostrar resultados.

```python
num_a = float(input("Informe o primeiro número: "))
num_b = float(input("Informe o segundo número: "))
```

```python
soma = num_a + num_b
print(f'{num_a:.2f} + {num_b:.2f} = {soma:.2f}')
subtracao = num_a - num_b
print(f'{num_a:.2f} - {num_b:.2f} = {soma:.2f}')
multiplicacao = num_a * num_b
print(f'{num_a:.2f} x {num_b:.2f} = {soma:.2f}')
divisao = (num_b != 0.0) and (num_a / num_b)
print(f'{num_a:.2f} / {num_b:.2f} = {soma:.2f}')
modulo = (num_b != 0.0) and (num_a % num_b)
print(f'{num_a:.2f} % {num_b:.2f} = {soma:.2f}')
potenciacao = num_a ** num_b
print(f'{num_a:.2f} ^ {num_b:.2f} = {soma:.2f}')
```
